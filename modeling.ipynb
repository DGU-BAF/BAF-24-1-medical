```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn.objects as so
import missingno as msno 
import os
from fancyimpute import IterativeImputer# Window의 한글 폰트 설정
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler
# import autosklearn.classification
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv('final_modeling_data.csv',encoding='cp949')
df_val = pd.read_csv('final_modeling_val_data.csv',encoding='cp949')
```

```python
# ***** 한 번만 하고 말아야됨 *****
df = df.drop(df.columns[[0, 1]], axis=1)
```

```python
# ***** 한 번만 하고 말아야됨 *****
df_val = df_val.drop(df_val.columns[[0, 1]], axis=1)
```

```python
X_train = df.drop('DIAG_NM', axis=1)  # 타겟 변수를 제외한 모든 열을 특성으로 선택
y_train = df['DIAG_NM']  # 타겟 변수로 'DIAG_NM'을 선택

X_val = df_val.drop('DIAG_NM', axis=1)
y_val = df_val['DIAG_NM']

# 의사 결정 트리 모델 생성
model = DecisionTreeClassifier()

# 모델 훈련
model.fit(X_train, y_train)

# 검증 데이터로 예측
y_pred = model.predict(X_val)

# 정확도 계산
accuracy = accuracy_score(y_val, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier

# 랜덤 포레스트 모델 생성
rf_model = RandomForestClassifier()

# 모델 훈련
rf_model.fit(X_train, y_train)

# 검증 데이터로 예측
y_pred_rf = rf_model.predict(X_val)

# 정확도 계산
accuracy_rf = accuracy_score(y_val, y_pred_rf)
print("Random Forest Accuracy:", accuracy_rf)
```

```python
from sklearn.linear_model import LogisticRegression

# 로지스틱 회귀 모델 생성
logistic_model = LogisticRegression()

# 모델 훈련
logistic_model.fit(X_train, y_train)

# 검증 데이터로 예측
y_pred_logistic = logistic_model.predict(X_val)

# 정확도 계산
accuracy_logistic = accuracy_score(y_val, y_pred_logistic)
print("Logistic Regression Accuracy:", accuracy_logistic)
```

```python
from sklearn.svm import SVC

# SVM 모델 생성
svm_model = SVC()

# 모델 훈련
svm_model.fit(X_train, y_train)

# 검증 데이터로 예측
y_pred_svm = svm_model.predict(X_val)

# 정확도 계산
accuracy_svm = accuracy_score(y_val, y_pred_svm)
print("SVM Accuracy:", accuracy_svm)
```

```python
from sklearn.metrics import confusion_matrix

# 혼동 행렬 계산
conf_matrix_svm = confusion_matrix(y_val, y_pred_svm)

# 혼동 행렬 출력
print("Confusion Matrix (SVM):")
print(conf_matrix_svm)
```
